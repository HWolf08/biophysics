```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

from numpy.random import default_rng
rng = default_rng()
rand = rng.random
```

```python
Illness_incidents = np.loadtxt("incidents.csv", delimiter=',')
```

```python
N = np.size(Illness_incidents,0)

X_Coord = Illness_incidents[0:N:1,0]
Y_Coord = Illness_incidents[0:N:1,1]
```

```python
plt.scatter(X_Coord, Y_Coord, c= "orange", marker= '.')
plt.title ('Coordinate Plane of Infections')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
```
