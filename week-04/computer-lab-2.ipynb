```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

from numpy.random import default_rng
rng = default_rng()
rand = rng.random
```

```python
num_steps = 1000
X_Step = 2*(rand(num_steps) < 0.5) - 1
print("X Step: ", X_Step)
Y_Step = 2*(rand(num_steps) < 0.5) - 1

print("Y Step: ", Y_Step)
```

```python
plt.plot (np.cumsum(X_Step, axis=None, dtype=None, out=None), np.cumsum(Y_Step, axis=None, dtype=None, out=None))
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Random Walk Path')
```

```python
NWalks = 100
num_steps = 1000
X_final = np.zeros(NWalks)
Y_final = np.zeros(NWalks)
Displacement = np.zeros(NWalks)

for n in range(NWalks):
    X = 2*(rand(num_steps) < 0.5) - 1
    Y = 2*(rand(num_steps) < 0.5) - 1
    X_Step = X.cumsum()
    Y_Step = Y.cumsum()

    X_final[n] = X_Step[-1]
    Y_final[n] = Y_Step[-1]
    Displacement[n] = np.sqrt(X_Step[-1]**2 + Y_Step[-1]**2)
```

```python
plt.scatter (X_final, Y_final)
plt.xlabel('X Original Location')
plt.ylabel('Displacement Distance from X Starting Point')
plt.title("Scatter Plot Detailing the \n Displacement Between the Start and End of 100 Random Walks")
```
