```python
import numpy as np
import matplotlib.pyplot as plt
import math

from scipy.integrate import odeint
```

```python
m = 1
g = 3
chi= 1
L = 1
o1 = 0
o2 = 2
o3 = 4
o4 = 6

theta1 = np.linspace(-np.pi/4,np.pi/4,101)
y = np.linspace(0, 10, 101)
X,Y= np.meshgrid(theta1,y)

theta2 = np.linspace(-np.pi/2,np.pi/2,101)
y = np.linspace(0, 10, 101)
X,Y= np.meshgrid(theta2,y)

theta3 = np.linspace(-np.pi,np.pi,101)
y = np.linspace(0, 10, 101)
X,Y= np.meshgrid(theta3,y)

theta4 = np.linspace(-np.pi*1.5,np.pi*1.5,101)
y = np.linspace(0, 10, 101)
X,Y= np.meshgrid(theta4,y)

o_time1 = -((m*g*np.sin(theta1)) + (chi*L*o1))/ (m*L)
o_time2 = -((m*g*np.sin(theta2)) + (chi*L*o2))/ (m*L)
o_time3 = -((m*g*np.sin(theta3)) + (chi*L*o3))/ (m*L)
o_time4 = -((m*g*np.sin(theta4)) + (chi*L*o4))/ (m*L)

x = np.linspace(-math.pi*3,math.pi*3, 20)
y = np.linspace(-math.pi*3,math.pi*3, 20)
X, Y = np.meshgrid(x,y)
vX = Y
vY = -X

Figure = plt.figure(figsize=(16,9))
plt.plot(theta1,o_time1, label= 'Theta = Quarter Pi', color='red')
plt.plot(theta2,o_time2,label= 'Theta = Half Pi', color='blue')
plt.plot(theta3,o_time3,label= 'Theta = Pi', color='orange')
plt.plot(theta4,o_time4,label= 'Theta = 3/2 Pi', color='Purple')
plt.quiver(X, Y, vX, vY, pivot='mid')
plt.legend(loc='upper right')

plt.axhline(y=0, color='red')
plt.plot(0,0,'og', ms=20)
```
