```python
import numpy as np
import matplotlib.pyplot as plt
import math

from scipy.integrate import odeint
```

## (a) Plot the phase portrait.

```python
m = 1
g = 3
chi= 1
L = 1
o = 0

theta = np.linspace(-np.pi,np.pi,101)
y = np.linspace(0, 10, 101)
X,Y= np.meshgrid(theta,y)

o_time = -((m*g*np.sin(theta)) + (chi*L*o))/ (m*L)
plt.plot(theta,o_time,'r')
```

```python
x = np.linspace(-math.pi,math.pi, 20)
y = np.linspace(-math.pi,math.pi, 20)
X, Y = np.meshgrid(x,y)
vX = Y
vY = -X
plt.quiver(X, Y, vX, vY, pivot='mid')

plt.plot(0,0,'og', ms=20)
plt.plot(-math.pi, 0, 'or', ms=20)
```

```python
Figure = plt.figure(figsize=(16,9))
plt.plot(theta,o_time,'r')
plt.quiver(X, Y, vX, vY, pivot='mid')

plt.axhline(y=0, color='red')
plt.plot(0,0,'og', ms=20)
plt.plot(-math.pi, 0, 'or', ms=20)
```
