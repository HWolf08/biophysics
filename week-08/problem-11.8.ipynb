## (a) Dimensional analysis

- What combination of $m$, $g$, and $\zeta$ defines the timescale $T$?

***Using the variable T from the original equation we found in class, T is equal to m/chi which has the dimensions of s^-1. Given this new variable, we can define a t_bar as equal to t/T so that it is dimensionless..***

- What does the differential equation look like when we rescale $t$ by $T$?

***When you rescale the problem with this new variable of t_bar, the equation becomes d^2pheta/dt_bar^2 = -(gT^2/L)*sin(theta)-(dtheta/dt_bar)***

- What does the differential equation look like when we make the small angle approximation $\sin \theta \approx \theta$?

***By replacing the sin(theta) just with theta due to the small angle approximation, the equation becomes d^2pheta/dt_bar^2 = -(gT^2/L)*(theta)-(dtheta/dt_bar). Which can be further reduced to d^2pheta/dt_bar^2 = -alpha*(theta)-(dtheta/dt_bar)***

- What is the solution of this approximate differential equation?

***Due to the characteristics of the second order derivative, the solution would approxiamtely be a sinusoidal wave.***

## (b) Classify solutions

What criterion determines whether solutions will overshoot the stable point or not?

```python
def F(y,t):
    dy = [0,0]
    dy[0] = y[1]
    dy[1] = -(alpha*y[0] + y[1])
    return dy
```

```python
m = 1
g = 9.81
L = 1
chi = 1

T = chi/m
alpha =(g*T**2)/L

t = np.linspace(0,10,31)
y0 = [1,4]
y = odeint(F,y0,t)

plt.plot(t,y[:,0])
plt.plot(t,y[:,1])
```

***By changing the chi value within the equation you can manipulate the entire functions response. This manipulates the dampening of the function that happens within the system.***
