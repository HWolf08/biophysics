```python
#5.1.2
HIV_data = np.loadtxt("HIVseries.csv", delimiter=',')
```

```python
N = np.size(HIV_data,0)

Time = HIV_data[0:N:1,0]
Concentration = HIV_data[0:N:1,1]
```

```python
plt.scatter(Time, Concentration, c= "orange", marker= 'x', label='Concentration vs. Days')
plt.title ('Infected HIV Concentration vs. Time in Days')
plt.xlabel('Time (Days)')
plt.ylabel('Concentration')
plt.legend(loc='upper right')
```

```python
#Assignment
time = np.linspace(0,7,101)
time = t

N = np.size(HIV_data,0)
Time = HIV_data[0:N:1,0]
Concentration = HIV_data[0:N:1,1]
```

```python
A= 90000
B= 90000
alpha = .65
beta = .35
```

```python
viral_load = A * np.exp(-alpha*t) + B * np.exp(-beta*t)

plt.plot(time, viral_load, c="yellow", label='Function data')
plt.scatter(Time, Concentration, c= "purple", marker= '+', label='HIV.csv Data')
plt.title ('Infected HIV Concentration vs. Time in Days')
plt.xlabel('Time (Days)')
plt.ylabel('Concentration')
plt.legend(loc='upper right')
```

```python
time = np.linspace(0,7,101)
time = t

N = np.size(HIV_data,0)
Time = HIV_data[0:N:1,0]
Concentration = HIV_data[0:N:1,1]

viral_load = A * np.exp(-(1/alpha)*t) + B * np.exp(-beta*t)

plt.plot(time, viral_load, c="yellow", label='Function data')
plt.scatter(Time, Concentration, c= "purple", marker= '+', label='HIV.csv Data')
plt.title ('Infected HIV Concentration vs. Time in Days')
plt.xlabel('Time (Days)')
plt.ylabel('Concentration')
plt.legend(loc='upper right')
```

```python
viral_load = A * np.exp(-alpha*t) + B * np.exp(-beta*t)
viral_load_2 = A * np.exp(-(1/alpha)*t) + B * np.exp(-beta*t)

plt.plot(time, viral_load, c="blue", label='alpha')
plt.plot(time, viral_load_2, c="red", label='inv. alpha')
plt.title ('Infected HIV Concentration vs. Time in Days')
plt.xlabel('Time (Days)')
plt.ylabel('Concentration')
plt.legend(loc='upper right')
```
