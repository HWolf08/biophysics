```python
import numpy as np 
import matplotlib.pyplot as plt

from numpy.random import default_rng
rng = default_rng()
rand = rng.random
```

## (a) __Histogram of $1/x$__

```python
x = np.random.uniform(0,1,10000)
Inv_x = 1/x
count, bins, ignored = plt.hist(Inv_x,50, density = True, range = (0,25))
```

## (b) Log-log plot of frequencies

```python
plt.loglog(count)
```

## (c) Analysis and Conclusions

The histogram of the uniform distribution indicates an exponential decay relationship. The equation for an exponential day includes f(x)= Ae^-kx, which when taken the the log of produes a negative linear relationship.This distribution makes sense and is familiar from what we have seen in prior evaluations.
