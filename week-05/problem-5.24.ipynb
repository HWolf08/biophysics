```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd

import scipy as sc
from scipy.stats import uniform,norm
from scipy.integrate import quad
```

## (a) Time-series and histogram

```python
df = pd.read_csv('djiweekly.csv',header=None)
df.columns = ['dji']

Week = np.arange(0,4224,1)
dji = df['dji'].to_numpy()
```

```python
u = dji[1:]/dji[:-1] 
y = np.log(u)

plt.figure(figsize = (16,9))
count,bins,ignore = plt.hist(y)
```

## (b) Gaussian approximation

```python
d_y = np.abs(bins[0]-bins[2])
u_y = 0

sigma = 0.05
Ntot = len(y)
gaussian = Ntot*(d_y/(sigma*np.sqrt(2*3.14)))*np.exp(-(y-u_y)**2/(2*sigma**2))

plt.plot(y,gaussian,"or")
count,bins,ignore = plt.hist(y)
```

## (c) Comparison of logarithms

```python
log_counts = np.log(count)
log_gaussian = np.log(gaussian)

plt.legend()
plt.plot(log_counts, label = 'log count')
plt.plot(log_gaussian, label = 'log gaussian')
```
