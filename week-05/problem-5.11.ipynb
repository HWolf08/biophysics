## (a) Sums of random variables

Write Python code to generate a histogram of 50,000 samples of a random variable $y = x_1 + x_2$ where $x_1$ and $x_2$ are random variables drawn from the uniform distribution over $[-1/2, 1/2]$.  Superimpose a plot of the expected probability distribution function for this variable.

```python
x1 = np.random.uniform(-.5,0.5,10000)
x2 = np.random.uniform(-.5,0.5,10000)
y = x1+x2

count, bins, ignored = plt.hist(y, 25, density=True)
count = np.insert(count,0,0)

plt.plot(bins,count, "-r")
```

## (b) Sums of _more_ random variables

Write Python code to generate a histogram of 50,000 samples of a random variable $y$ that is the sum of __10__ random variables drawn from the uniform distribution over $[-1/2, 1/2]$.  Superimpose a plot of the expected probability distribution function for this variable.

```python
Samples = 50000
x_1 = -.5
x_2 = .5

y = np.zeros(Samples)
x = np.zeros(10)
for j in range(Samples):
    for i in range(10):
        x[i] = np.random.uniform(x_1,x_2,1)
    y[j] = x.sum()

count, bins, ignored = plt.hist(y, 50,density=True)
count = np.insert(count,0,0)

plt.plot(bins,count, "-p")
```

```python
plt.semilogx(bins, count, marker = ".",
             markersize = 10,
             color = "green")
plt.show()
```

## (c) Sums of nonuniform discrete random variables

rite Python code to generate a histogram of 50,000 samples of a random variables $y$ that are the sum of __2__, __4__, and __10__ random variables drawn from the discrete distribution
$$
P(1) = \dfrac{1}{3} \qquad P(2) = \dfrac{2}{9} \qquad P(3) = \dfrac{1}{9} \qquad P(4) = \dfrac{1}{3}
$$

```python
def discrete_dist(value,probability,num_samples,num_sum):
    y = np.zeros(num_samples)
    for j in range(num_samples):
        y[j] = np.random.choice(value,num_sum,probability).sum()
    return y

discrete = [1,2,3,4]
prob = [1/3,2/9,1/9,1/3]

num_sum_tot = 4
num_samples_tot = 50000

x_4 = discrete_dist(discrete,prob,num_samples_tot,num_sum_tot)
bins_x4 = len(np.unique(x_4))
```

```python
num_sum_tot = 8
x_8 = discrete_dist(discrete, prob,num_samples_tot,num_sum_tot)
bins_x8 = len(np.unique(x_8))
```

```python
num_sum_tot = 12
x_12 = discrete_dist(discrete, prob,num_samples_tot,num_sum_tot)
bins_x12 = len(np.unique(x_12))
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,3,1).hist(x_4,bins=bins_x4)
plt.title('N = 4 Histogram')
plt.subplot(1,3,2).hist(x_8,bins=bins_x8)
plt.title('N = 8 Histogram')
plt.subplot(1,3,3).hist(x_12,bins=bins_x12)
plt.title('N = 12 Histogram')
```

## (d) Sums of nonuniform continuous random variables

Write Python code to generate a histogram of 50,000 samples of a random variables $y$ that are the sum of __5__ and __50__ random variables drawn from the exponential distribution with expectation 1.

```python
num_samples = 50000
y = np.zeros(num_samples)

for i in range(num_samples):
    y[i]=np.random.exponential(1,5).sum()
    
plt.hist(y,bins=25)
```

```python
num_samples = 50000
y = np.zeros(num_samples)

for i in range(num_samples):
    y[i]=np.random.exponential(1,50).sum()
    
plt.hist(y,bins=25)
```
