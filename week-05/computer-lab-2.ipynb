```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

from numpy.random import default_rng
rng = default_rng()
rand = rng.random

import scipy.special as sc
from scipy.special import factorial
```

```python
P = 0.08
l = np.arange(0,15,1)
P_l = ((np.exp(-8) * 8**l) / sc.factorial(l))
print(P_l)
```

```python
N = 100
num_flips = 100
M = np.zeros(N)

for n in range(N):
    flip = 2*(rand(num_flips) < 0.08) - 1 
    heads_count = (flip==1)
    count = heads_count.sum()
    M[n] = count
print(count)
```

```python
N = 1000
num_flips = 100
M = np.zeros(N)

for n in range(N):
    flip = 2*(rand(num_flips) < 0.08) - 1 
    heads_count = (flip==1)
    count = heads_count.sum()
    M[n] = count
print(count)
```

```python
plt.hist(M)
plt.xlabel('Number of Bins')
plt.ylabel('Number of Heads in Trial')
plt.title("Distribution of Heads for N Trials at 1,000")
```

```python
plt.hist(M)

N = 1000
l = np.arange(0,15,1)
P_l = (np.exp(-8) * 8**l / sc.factorial(l)) * N
plt.plot(P_l)
```

```python
N = 1000000
num_flips = 100
M = np.zeros(N)

for n in range(N):
    flip = 2*(rand(num_flips) < 0.08) - 1 
    heads_count = (flip==1)
    count = heads_count.sum()
    M[n] = count

plt.hist(M)
plt.xlabel('Number of Bins')
plt.ylabel('Number of Heads in Trial')
plt.title("Distribution of Heads for N Trials at 1,000,000")

l = np.arange(0,21,1)
P_l = (np.exp(-8) * 8**l / sc.factorial(l)) * N
plt.plot(P_l)
```

```python
#For the above graph, when compared to the previous graph of lower N count, there is a similar distribution. In addition to the distribution similarity, the line curve fits the distribution similarly. From reading the graphic, it makes sense as the most repeated number from the arrangement matches with the expected quantity.
```

```python
num_flips = 1000 
flips = (rand(num_flips) < 0.08) - 1 + 1
One = np.nonzero(flips)
Dif_One = np.diff(One).flatten()

counts = np.unique(Dif_One,return_counts=True)
```

```python
plt.bar(counts[0],counts[1])
plt.title('Frequency of Waiting Time')
plt.ylabel('Frequency')
plt.xlabel('Time')
```

```python
#The resulting graphic for the waiting times provides a distribution that is relatively rough for seeing the characteristics. To remove all of the large dips that cause this character, I completed the below graphic.
```

```python
plt.hist(Dif_One,bins=10)
plt.title('Frequency of Waiting Times')
plt.ylabel('Frequency')
plt.xlabel('Time')
```
