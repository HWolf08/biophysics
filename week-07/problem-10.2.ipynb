## (a)

```python
data = np.load('colquhounData.npy')
N_chan = data.sum()

print(N_chan)
```

```python
divide = np.hsplit(data,2)
sections = divide[0].flatten()
N_chan2 = divide[1].flatten()

print(N_chan2)
```

```python
chan_open = N_chan2.cumsum()

print(chan_open)
```

## (b)

```python
plt.figure()
plt.plot(sections, N_chan2)
plt.title('Number of Closed Channels')
```

```python
plt.figure()
plt.plot(sections, chan_open)
plt.title('Number of Open Channels')
```
